```python
# Iteration of list
# Iteration by index

# Index = 0 to 9
# How many elements? = 10
lst = [67, 54, 11, 99, 98, 543, 432, "Anirudh", True, 55.556]

# Best
for i in range(len(lst) - 1, -1, -1):
    print(lst[i], end=" ")
```

```python
lst = [67, 54, 11, 99, 98, 543, 432, "Anirudh", True, 55.556]

for i in range(-1, -(len(lst) + 1), -1):
    print(lst[i], end=" ")
```

```python
lst = [67, 54, 11, 99, 98, 543, 432, "Anirudh", True, 55.556]
for i in range(0, len(lst)):
    print(lst[i], end=" ")
```

```python
my_list = [67, 54, 11, 99, 98, 543, 432]

# Addition of all elements using iteration

total=0
for i in my_list:
    total+=i

total
```

```python
# Membership Operator (IN / NOT IN)

my_list: list[int] = [54, 99, 18, 19, 99, 21, 59, True]

search_number = 99

if search_number in my_list:
    print("Yes")
else:
    print("No")
```

```python
#Removes an element
#AND returns the removed value
my_list:list[int]= [54, 99, 18, 19, 99, 21, 59]
my_list.pop(-3)
my_list
```
