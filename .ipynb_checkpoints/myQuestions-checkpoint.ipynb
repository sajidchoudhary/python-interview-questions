```python
"""
Sort dict without any build-in function
"""
my_dict = {"apple": 6, "banana": 3, "pear": 10, "orange": 5}

sorted_dict = {}

while my_dict:
    min_key = None
    min_value = None #Use None so the first value automatically becomes the minimum.

    for key, value in my_dict.items():
        if min_value is None or value < min_value:
            min_value = value
            min_key = key

    sorted_dict[min_key] = min_value
    del my_dict[min_key]

print(sorted_dict)
```

```python
"""
Sort dict without any build-in function by key
"""
my_dict = {"apple": 6, "banana": 3, "pear": 10, "orange": 5}
sorted_dict = {}
while my_dict:
    min_key = None
    for key in my_dict:
        if min_key is None or key < min_key:
            min_key = key
    sorted_dict[min_key] = my_dict[min_key]
    del my_dict[min_key]

print(sorted_dict)
```

```python
'''
sort the dictionary by the length of the keys without any build-in function.
'''

my_dict = {"apple": 6, "banana": 3, "pear": 10, "orange": 5}
sorted_dict={}
while my_dict:
    min_key=None
    for key in my_dict:
        if min_key is None or len(key)<len(min_key):
            min_key=key
    sorted_dict[min_key]=my_dict[min_key]
    del my_dict[min_key]

sorted_dict
```
