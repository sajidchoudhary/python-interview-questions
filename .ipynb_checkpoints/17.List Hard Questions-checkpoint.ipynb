```python
"""
Write a Python code to check if a given list is sorted in ascending order.
"""
#Method1
lst=[14,15,16,1,17,18]
is_sorted=True

for i in range(1,len(lst)):
    if lst[i]<lst[i-1]:
        is_sorted=False
        break
        
print(is_sorted)  



#Method2:
def isSorted(my_list:List)->bool:
    for i in range(0, len(my_list) - 1):
        if my_list[i] > my_list[i + 1]:
            return False
    return True

lst=[14,15,16,17,18]
isSorted(lst)
```

```python
"""
Given two sorted lists, write a Python code to merge them into a single sorted list.
"""
lst1=[14,15,16,17,18]
lst2=[20,21,22,23,24,25]

#Method1:
def twoListSort(lst1:List, lst2:List)->List:
    singleList=lst1+lst2
    newList=[]
    largest=singleList[0]

    for i in range(len(singleList)):
        if singleList[i]>largest:
            newList.append(largest)
            largest=singleList[i]
    return newList


newList= twoListSort(lst1,lst2)  
print(newList)


#Method2
a = [1, 4, 6, 7, 8, 12, 14, 22]
b = [34, 56, 78, 99, 100]

i = 0
j = 0

result = []

while i < len(a) and j < len(b):
    if a[i] < b[j]:
        result.append(a[i])
        i += 1
    elif b[j] < a[i]:
        result.append(b[j])
        j += 1
    else:
        result.append(a[i])
        result.append(b[j])
        i += 1
        j += 1

while i < len(a):
    result.append(a[i])
    i += 1

while j < len(b):
    result.append(b[j])
    j += 1

print(result)
```

```python
"""
Write a python program to check if the list contains three consecutive common numbers in Python.
"""

#Method1:
abc=[1,2,3,4,4,4,4,5,6,7,7,8,9,9,9,9,9,10]
def consecutive_common_numbers(abc):
    cnt=0
    newList=[]
    for i in range(1,len(abc)):
        if abc[i]==abc[i-1]:
            cnt+=1
    
        if cnt==3:
            newList.append(abc[i-1])
            cnt=0
    return newList

myList=consecutive_common_numbers(abc)
print(myList)


def isConsecutive(myList):
    for i in range(0,len(myList)-2):
        if myList[i]==myList[i+1]==myList[i+2]:
            print(myList[i])

isConsecutive(abc)
```

```python
"""
Write a Python code to check if a list is a palindrome 
(reads the same forwards and backwards). Just print “YES” or “NO”
"""

def isPalindrome(lst: List) -> bool:
    for i in range(0, len(lst)):
        if lst[i] != lst[len(lst) - i - 1]:
            return False
    return True

print(isPalindrome([6,1, 2, 3, 4, 4, 3, 2, 1,6]))
print(isPalindrome([1, 2, 3, 4, 3, 2, 1]))
print(isPalindrome([1, 2, 3, 4, 3, 4, 1]))
```

```python
"""
Write a Python code to find the occurrence of each element in a list 
and print the element with the highest occurrence.
"""

my_list = ["sajid", 1, 1, 1, 2, 3, 154, 65, 2, 2, 1, 44, 5, 154, 154, "sajid"]

unique_elements = []

for i in my_list:
    if i not in unique_elements:
        unique_elements.append(i)

# Print the frequencies and keep track of the largest frequency and that element
highest_frequency = 0
highest_frequency_element = 0

for i in unique_elements:
    freq = my_list.count(i)
    print(f"{i} occurs {freq} times")
    if freq > highest_frequency:
        highest_frequency = freq
        highest_frequency_element = i

print(f"Highest frequency = {highest_frequency}")
print(f"Highest frequency element = {highest_frequency_element}")
```

```python
"""
Write a python code to find second largest element in list without sorting.
"""

def secondLargest(nums: List[int]):
    if len(nums) < 2:
        return "Not enough elements"

    largest = second_largest = float("-inf")  # -inf means the most negative value
    # Can take as the smallest value

    for num in nums:
        if num > largest:
            second_largest = largest
            largest = num
        elif num > second_largest and num != largest:
            second_largest = num

    if second_largest == float("-inf"):
        return "There is no second largest element"
    else:
        return second_largest


nums = [12, 74, -89, 96, -98, -96, 52]
result = secondLargest(nums)
print(f"Second largest element = {result}")
        
        
```
