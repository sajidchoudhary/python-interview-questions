```python
from typing import List
```

```python
"""
Create a function named oddCharacters which takes a string as a parameter. 
Now return a list of characters which appears odd times in that string.
"""
string="aeroplane"

#Method1:
def oddCharacters(string:str):
    count={}
    output=""
    for i in range(len(string)):
        if string[i] not in count:
            count[string[i]]=1
        else:
            count[string[i]]+=1
    for key, val in count.items():
        if val%2!=0:
            output+=key
    return output
            
output=oddCharacters(string)
print(output)

#Method2:
def oddCharacters(string: str) -> List[str]:
    characters_count = {}
    for ch in string:
        characters_count[ch] = characters_count.get(ch, 0) + 1
    result = []
    for char, count in characters_count.items():
        if count % 2 == 1:
            result.append(char)
    return result

print(oddCharacters("aeroplane"))
```

```python
"""
Create a function named arrangeChars which takes a string as a parameter. 
Now return a string with max frequency chars at start.
"""
string="abcdeabcdabcabaxyz"

def arrangeChars(string):
    output=""
    characters_count={}
    for i in string:
        characters_count[i]=characters_count.get(i,0)+1

    sorted_characters_count=dict(sorted(characters_count.items(), key=lambda x:x[1], reverse=True))
    for key,val in sorted_characters_count.items():
        output+=key*val
    return output
      
output=arrangeChars(string)
print(output)
```

```python
"""
Given a string S, containing numeric words, the task is to 
convert the given string to the actual number.
"""
help_dict = {
    "one": "1",
    "two": "2",
    "three": "3",
    "four": "4",
    "five": "5",
    "six": "6",
    "seven": "7",
    "eight": "8",
    "nine": "9",
    "zero": "0",
}

string="zero four zero one"
#Method1:
words=string.split()
output=""
for i in words:
    if i in help_dict:
        output+=help_dict[i]
print(output)

#method2:
res = "".join(help_dict[ele] for ele in string.split())
print(res)
```

```python
"""
Write a Python program to capitalize the first and last 
letters of each word in a given string.
"""

name="sajid choudhary"

#Method1:
def capitalize_first_last_letters1(str1):
    str1 = result = str1.title()
    result = ""
    for word in str1.split():
        result += word[:-1] + word[-1].upper() + " "
    return result[:-1]

result=capitalize_first_last_letters1(name)
print(result)

#Method2:
def capitalize_first_last_letters2(sentence):
    result=""
    words=sentence.split()
    for word in words:
        result+=word[0].upper()+word[1:-1]+word[-1].upper()
        result+=" "
        
    return result

result1=capitalize_first_last_letters2(name)
print(result1)

print(capitalize_first_last_letters1("python is a great language"))
print(capitalize_first_last_letters2("delhi is best city with 0 AQI"))
```

```python
"""
Write a Python program to generate two strings from a given string.
for the first string, use the characters that occur only once, and for the second,
use the characters that occur multiple times in said string.
Input=aabbcceffgh
Output:
string1: egh
string2: abcf
"""
## Method1:

def twoString1(input):
    char_freq={}
    str1=""
    str2=""
    for i in input:
        char_freq[i]=char_freq.get(i, 0)+1
    for k, v in char_freq.items():
        if v!=1:
            str2+=k
        else:
            str1+=k
    return str1, str2


def twoString2(input):
    char_freq={}
    for i in input:
        char_freq[i]=char_freq.get(i, 0)+1
    part1="".join(k for k,v in char_freq.items() if v==1)
    part2="".join(k for k,v in char_freq.items() if v!=1)
    return part1, part2



input="aabbcceffgh"
str1, str2 = twoString(input)
print(str1)
print(str2)

part1, part2 = twoString2(input)
print(part1)
print(part2)
```
