```python
# Sets
my_set = {1, 2, 3, 4, "Sajid", 4, 4, "Sajid", "x", "y", 2}

print(my_set)
print(type(my_set))

# It is iterable
for i in my_set:
    print(i)

# Slicing not allowed coz no index
print(my_set[:2])
```

```python
#empty set creation
my_set = set()

print(my_set)
print(type(my_set))


x = float()

print(x, type(x))
```

```python
a = {1, 2, 3, 4}
print(type(a))
```

```python
#Operators in Set
a = {1, 2, 3, 4, 5}
b = {3, 4, 5, 6, 7, 8}

print(a + b)
print(a - b)
print(a / b)
print(a * b)
```

```python
#Sets methods

a = {1, 2, 3, 4, 5, "Sajid", 66.77, (1, 2, 3)}
b = {3, 4, 5, 6, 7, 8}

print(a)
a.clear()
print(a)
```

```python
a = {1, 2, 3, 4, 5, "Sajid", 66.77, (1, 2, 3)}
a.add(100)
print(a)

a.pop()  # Remove from first
print(a)

a.remove(500)
print(a)
```

```python
a = {1, 2, 3, 4, 5}
b = {4, 5, 6, 7, 8}
c = {4, 5, 6, 7, 8}
d = {4, 5, 6, 7, 8, 100, 200}
#common from all sets
x = a | b | c | d
print(x)
```

```python
x = a.copy()

x.remove(5)
print(x)
print(a)

print(id(a))
print(id(x))
```

```python
a = {1, 2, 3, 4, 5, "Sajid", 66.77}
b = {3, 4, 5, 6, 7, 8}

x = a.union(b)  # All elements from both sets
print(x)
```

```python
x = a.intersection(b)  # Common elements
print(x)

x = a.difference(b)
print(x)
```

```python
x = b.difference(a)
print(x)
```
