```python
'''
Given an integer array nums and an integer k, 
return the k most frequent elements within the array. 
You may return the output in any order. 
Example 1:Input: nums = [1,2,2,3,3,3], k = 2 Output: [2,3] 
Input: nums = [7,7], k = 1 Output: [7]
'''

nums = [5,5,5,5,1,2,2,3,3,3,3]
k = 2
def topKFrequent(nums,k):
    freq={}
    for i in range(len(nums)):
        if nums[i] not in freq:
            freq[nums[i]]=1
        else:
            freq[nums[i]]+=1
    
    topk=[key for key, value in sorted(freq.items(), key=lambda x:x[1], reverse=True)[:k]]
    return topk
output=topKFrequent(nums,k)
output
```

```python
doc = {
"title": "Company Leave Policy",
"text": """
Employees receive 20 days of annual leave. Sick leave policies allow up to 10 days per year.
Leave requests must be submitted through the HR portal and approved by the reporting manager.
Extended leave beyond 30 days requires HR approval.
"""
}

"8 words ke chunks chahiye aur 2 words overlap bhi chahiye, saath hi metadata (id, title) add karna hai."

chunk_size = 8
overlap = 2
step = chunk_size - overlap

words = doc["text"].split()

chunks = []

for i in range(0, len(words), step):
    chunk_words = words[i:i + chunk_size]
    
    if len(chunk_words) < chunk_size:
        break
        
    chunk = {
        "id": len(chunks) + 1,
        "title": doc["title"],
        "text": " ".join(chunk_words)
    }
    
    chunks.append(chunk)

for c in chunks:
    print(c)
```

```python
'''
Given a string s, return a string that contains each character along with its frequency in the input string.
The output should follow these rules:
Characters must be sorted in descending order of frequency.
If two or more characters have the same frequency, sort them in alphabetical order.
Append each character followed by its frequency to form the final string.

Examples
Example 1:
Input: s = "ggggoolaaanng"
Output: "g5a3n2o2l1"

Example 2:
Input: s = "totally"
Output: "l2t2a1o1y1"
'''


def patternPrint(input: str) -> str:
    freq_count = {}
    # Step 1: Count frequency
    for ch in input:
        freq_count[ch] = freq_count.get(ch, 0) + 1
    
    # Step 2: Sort by (-frequency, character)
    sorted_items = sorted(freq_count.items(), key=lambda x: (-x[1], x[0]))
    
    # Step 3: Build output string
    output = ""
    for ch, freq in sorted_items:
        output += ch + str(freq)
    return output

print(patternPrint("ggggoolaaanng"))
```

```python
#input
car_list = [
    {'Brand': 'Hyundai', 'Model': 'CRETA', 'Year': 2021, 'Colors': ['RED', 'BLUE', 'WHITE', 'BLACK']},
    {'Brand': 'Maruthi', 'Model': 'SWIFT', 'Year': 2021, 'Colors': ['RED', 'BLUE', 'WHITE','GREEN']},
    {'Brand': 'TATA', 'Model': 'PUNCH', 'Year': 2021, 'Colors': ['ORANGE', 'BLUE', 'WHITE', 'GREEN']}
]
#Output: Get me the list of cars with GREEN Color
#['Maruthi', 'TATA']


output=[]
for car in car_list:
    if "GREEN" in car['Colors']:
        output.append(car['Brand'])
        
print(output)        
        
#using list comprehension        
green_cars = [car['Brand'] for car in car_list if 'GREEN' in car['Colors']]
print(green_cars)
```

```python
# convert Two lists of random lengths into dictonary.
keys = ["apple", "banana", "cherry"]
values = [1, 2, 3, 4]

# Create a dictionary from the lists
my_dict = {}
for i in range(min(len(keys),len(values))):
    my_dict[keys[i]] = values[i]

print(my_dict)


# Two lists of random lengths
keys = ["apple", "banana", "cherry"]
values = [1, 2, 3, 4]

# Create a dictionary using zip
my_dict = dict(zip(keys, values))

print(my_dict)
```

```python
#if input is "aaaabbcddd" then output should be "a4b2c1d3" OR aaaabbcdddaa = a4b2c1d3a2
def compress_string(s):
    result = ""
    count = 1
    for i in range(1, len(s)):  #if range start from 0 then [i-1]=[0-1]=[-1] toh last wala element sabse phle aa jaye ga.
        if s[i] == s[i - 1]:  #jab tk previous char aur current char same hai tb tk count me add hoga
            count += 1
        else:
            result += s[i - 1] + str(count) #jab previous char aur current char alag honge tab previous char me count ka final count add kar denge
            count = 1

    result += s[-1] + str(count) #last character loop ke ander compare nahi ho pata iss liye uss ko loop se bahar aa kar final count karte hai
    return result

# Example usage:
input_str = "AAAARRJJJKKO"
output_str = compress_string(input_str)
print(output_str)

#Note: run this in new ipynb file
```
