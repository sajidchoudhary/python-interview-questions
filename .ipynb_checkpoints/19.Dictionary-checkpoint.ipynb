```python
student_details = {
    "sajid": {
        "age": 66,
        "gender": "Male",
        "address": "mzn",
        "phone": 100,
        "physics": 98,
        "chemistry": 87,
        "maths": 11,
    },
    "nihar": {
        "phone": 3213421,
        "gender": "Male",
        "address": "Delhi",
        "age": 16,
        "physics": 54,
        "chemistry": 71,
        "maths": 91,
    },
    "anusha": {
        "phone": 585652,
        "gender": "Female",
        "address": "Agra",
        "age": 52,
        "physics": 14,
        "chemistry": 64,
        "maths": 71,
    },
}


for name, details in student_details.items():
    total = (
        details.get("physics", 0)
        + details.get("chemistry", 0)
        + details.get("maths", 0)
    )
    details["total_marks"] = total

print(student_details)
```

```python
details = {
    "sajid": [78, 43, 22],
    "arshi": [11, 56, 23],
    "muskan": [98, 91, 3],
}

for k, v in details.items():
    total=0
    for i in v:
        total=total+i
    print(f"{k} has {total} marks")
```

```python
highest_marks=0
highest_marks_student_name=0

for k,values in details.items():
    if sum(values)>highest_marks:
        highest_marks=sum(values)
        highest_marks_student_name=k

print(highest_marks)
print(highest_marks_student_name)
```

```python
#without using sum()
for k, values in details.items():
    highest_marks=0
    for i in values:
        if i>highest_marks:
            highest_marks=i
    print(f"{k} has scored highest marks {highest_marks}")
```

```python
def convert(n: int):
    if n % 2 == 0:
        return n + 1
    return n - 1


my_list = [34, 54, 17, 65]

x = list(map(convert, my_list))
print(x)
```

```python
def convert(n:int):
    if n%2!=0:
        return n-1
    return n+1

my_list = [34, 54, 17, 65]
list(map(convert, my_list))
```

```python
"""
Ask how many students = 2

Enter student name = sajid
Enter marks 1 = 56
Enter marks 2 = 13
Enter marks 3 = 98
Enter student name = arshi
Enter marks 1 = 56
Enter marks 2 = 13
Enter marks 3 = 98
"""
myDict={}

noOfStudents=int(input("Enter no of students:"))

for _ in range(noOfStudents):
    name=str(input("Enter name of student:"))
    noOfSubs=int(input("Enter no of subjects:"))
    marks=[]
    for i in range(noOfSubs):
        mark=int(input("Enter mark:"))
        marks.append(mark)
    myDict[name]=marks
print(myDict)

```
