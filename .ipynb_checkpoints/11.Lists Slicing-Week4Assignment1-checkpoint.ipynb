```python
from typing import List
```

```python
"""Make your own list of number. Ask a start and end position from User. 
Print the list from start position to end position using slicing."""

def listSlice(lst: List, start: int, end: int):
    print(lst[start : end + 1])


my_list = ["Anirudh", 6, 4, 110.654, True, -54]
listSlice(my_list, 2, 4)
```

```python
'''
Make your own list. write a python program that takes an integer as an input, 
and make a new list containing the last n elements of the original list. Using slicing logic
'''

def listLastNSlice(lst: List, n: int):
    l = len(lst)
    print(lst[l - n :])


my_list = ["Anirudh", 6, 4, 110.654, True, -54]
listLastNSlice(my_list, 2)
```

```python
'''Write a python program to interchange first and last elements in a list'''
def interchange(lst: List):
    # lst[0], lst[-1] = lst[-1] = lst[0]  # Method 1

    # Method 2
    temp = lst[0]
    lst[0] = lst[-1]
    lst[-1] = temp


my_list = ["Anirudh", 6, 4, 110.654, True, -54]
interchange(my_list)
print(my_list)
```

```python
'''
Write a python code to split a list into two halves using list slicing. (keep the length of list even).
'''
def splitList(lst: List):
    mid: int = len(lst) // 2
    first_half = lst[:mid]
    second_half = lst[mid:]
    print(first_half)
    print(second_half)


my_list = ["Anirudh", 6, 4, 110.654, True, -54]
splitList(my_list)
```
