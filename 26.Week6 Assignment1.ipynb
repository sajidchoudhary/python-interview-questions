```python
'''
Here is the Student Data and their marks.
student_data = {
    "Alice": [85, 90, 88, 92, 89],
    "Bob": [78, 82, 79, 81, 80],
    "Charlie": [92, 95, 88, 85, 91],
    "Diana": [76, 80, 79, 82, 85],
    "Eva": [88, 92, 85, 90, 87],
    "Frank": [83, 85, 80, 86, 88],
    "Gina": [90, 87, 92, 88, 86],
}
Display the name of student and total marks in asending order.
'''

student_data = {
    "Alice": [85, 90, 88, 92, 89],
    "Bob": [78, 82, 79, 81, 80],
    "Charlie": [92, 95, 88, 85, 91],
    "Diana": [76, 80, 79, 82, 85],
    "Eva": [88, 92, 85, 90, 87],
    "Frank": [83, 85, 80, 86, 88],
    "Gina": [90, 87, 92, 88, 86],
}

#Method1:

newDict={}
for stu, marks in student_data.items():
    total=0
    for m in marks:
        total+=m
    newDict[stu]=total

final_output=dict(sorted(newDict.items(), key=lambda x:x[1], reverse=False))
print(final_output)

sorted_student_data = dict(sorted(student_data.items(), key=lambda x: sum(x[1])))  #sorted by marks in asc
for name, marks in sorted_student_data.items():
    total_marks = sum(marks)
    print(f"{name} has scored {total_marks}")
```
