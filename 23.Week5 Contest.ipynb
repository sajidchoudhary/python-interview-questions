```python
"""
Given a list of strings, concatenate them into a single string separated by spaces.
For example, given the input ["Hello", "World", "Python"], 
the output should be "Hello World Python".

Make a list on your own.Don't use the JOIN function
"""

from typing import List


def joinWords(words: List[str]) -> str:
    result = ""
    for word in words:
        result += word
        result += " "
    return result


my_words = ["Hello", "World", "Python"]

s = joinWords(my_words)
print(s)
```

```python
"""
Write a Python program to convert a given string to all uppercase 
if it contains at least 2 uppercase characters in the first 4 characters.
input=pyTHon
output=PYTHON

input= helLo
output=helLo

input= gOOd
"""

#Method1:
def convertToUppercase1(s: str) -> str:
    upper_count = sum(1 for c in s[:4] if c.isupper())
    if upper_count >= 2:
        return s.upper()
    return s


def convertToUppercase2(s: str):
    upper_count = 0
    for c in s[:4]:
        if c.isupper():
            upper_count += 1
    if upper_count >= 2:
        return s.upper()
    return s

result = convertToUppercase1("HelLo")
print(result)

result = convertToUppercase2("HelLo")
print(result)
```

```python
"""
Create a dictionary that counts the frequency of words in a given string. 
Ask string from user.
"""

#Method1:
def countFreq():
    freq={}
    userInput=str(input("Enter sentence:"))
    splitSentence=userInput.split()
    for i in range(len(splitSentence)):
        if splitSentence[i] not in freq:
            freq[splitSentence[i]]=1
        else:
            freq[splitSentence[i]]+=1
    return freq

output=countFreq()
print(output)

#Method2:

def wordFrequency(string: str):
    words = string.split()
    frequency_dict = {}
    for word in words:
        frequency_dict[word] = frequency_dict.get(word, 0) + 1
    return frequency_dict
        
frequency = wordFrequency("python is is a good python good python")
print(frequency)
```

```python
'''
Write a Python Program to map two lists into a dictionary.
Everything in both lists  should be unique.
Example1:
list1=['red','green','blue']
list2=['FA8000#', '#78POI00', '0#HGAS745']
output={'red':'FA8000#', 'green':'#78POI00', 'blue':'0#HGAS745'}
'''

list1=['red','green','blue']
list2=['FA8000#', '#78POI00', '0#HGAS745']

#Method1:
output1=dict(zip(list1, list2))
print(output1)

#Method2:
output2={}
for i in range(len(list1)):
    output2[list1[i]]=list2[i]
    
output2
```

```python
'''
Write a python program to convert string values of a given dictionary into integer/float data type.
Example1:
original_list = [{'x':'10', 'y':'20', 'z':'30'},{'p':'40', 'q':'50', 'r':'60'}]
Output = [{'x':10, 'y':20, 'z':30},{'p':40, 'q':50, 'r':60}]

Example2:
original_list = [{'x':'10.36', 'y':'20.00', 'z':'30.98'},{'p':'40.41', 'q':'50.00', 'r':'60.00'}]
Output = [{'x':10.36, 'y':20.00, 'z':30.98},{'p':40.41, 'q':50.00, 'r':60.00}]
'''

def convertValues(data: list[dict]):
    for d in data:
        for key, value in d.items():
            if "." in value:
                d[key] = float(value)
            else:
                d[key] = int(value)
    return data


list1 = [
    {"x": "10", "y": "20", "z": "30"},
    {"p": "40", "q": "50", "r": "60"},
]

output1 = convertValues(list1)
print(output1)

list2 = [
    {"x": "10.12", "y": "20.23", "z": "30"},
    {"p": "40.00", "q": "50.19", "r": "60.99"},
]

output2 = convertValues(list2)
print(output2)
```
