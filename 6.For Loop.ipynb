```python
# 5 to 17 | -10 10

def printPattern(start, end):
    if start<end:
        for i in range(start, end+1):
            print(i, end=" ")
    else:
        for i in range(end, start+1):
            print(i, end=" ")


start=5
end=-15
printPattern(start, end)
```
