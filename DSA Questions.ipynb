```python
'''You are given a list of time intervals. Each interval has a start point and an end point. 
Two intervals are considered non-overlapping if one ends before (or exactly when) the other begins.
Find the largest number of non-overlapping intervals you can choose.
Example 1
Input:
l = [3, 4, 5]
r = [5, 9, 8]

Output:
2

Explanation:
An optimal selection of intervals is:
[3,5] and [5,8]

Example 2
Input:
l = [3, 5, 7]
r = [4, 6, 8]

Output:
3

Explanation:
An optimal selection of intervals is:
[3,4], [5,6], [7,8]

Constraints
1 ≤ size of l and r ≤ 2 × 10⁵'''

#My Explanation:
'''Intervals diye gaye hain. Interval matlab start time se end time tak ka time.
l = [3,4,5], r = [5,9,8]
Iska matlab intervals hain:
[3,5]
[4,9]
[5,8]
Format: [start, end]

Rule: Maximum intervals choose karne hain lekin wo overlap nahi hone chahiye.
Overlap matlab: [3,5]  aur  [4,9] Yeh overlap kar rahe hain kyunki 4,  3 aur 5 ke beech mein hai.

Step 1: First interval lo [3,5]
Step 2: Next interval check karo [4,9] Yeh overlap karta hai kyunki 4 < 5 Toh yeh nahi le sakte.
Step 3: Next interval [5,8] Yeh valid hai kyunki 5 >= 5 Matlab pehla interval khatam hua aur dusra start ho gaya.

Final selection:
[3,5]
[5,8]

Total intervals: 2
'''


def maxNonOverlapping(l, r):
    intervals = sorted(zip(l, r), key=lambda x: x[1])
    count = 0
    end = float('-inf')
    
    for start, finish in intervals:
        if start >= end:
            count += 1
            end = finish
            
    return count

l = [3,4,5]
r = [5,9,8]

output=maxNonOverlapping(l, r)
output


#agar interval lists return karni ho
def maxNonOverlapping(l,r):
    intervals=sorted(zip(l,r), key=lambda x:x[1])
    selected=[]
    end=float('-inf')
    for start, finish in intervals:
        if start>=end:
            selected.append([start, finish])
            end=finish
    return selected


l = [3,4,5]
r = [5,9,8]

output=maxNonOverlapping(l, r)
output
```
